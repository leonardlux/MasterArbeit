```python
import os
import numpy as np

from tools.file import read_config, write_config, get_standard_config, read_folder
from tools.combined import generate_new_data_from_config_file

from tools.parameter import PATH_TO_CONFIG_FOLDER
```


## Config Definitions and Data Generation

Steane circ (order 0p) with multiple QEC rounds, only Z observable 

Circuit level noise with with fixed noise rate 

Decoding both with ML and MWPM

```python
# Main parameters:
num_shots_ML = 1_000
# 104 min ?!
num_shots_MWPM = 10_000
# 10.3 s ?!

noise_rates = [0.001] 
distances = [3,5,7,9,11,13,15]
qec_rounds = [1,2,3,4,5,6]
```

```python
# ML Config 
folder_name_ML = "multi_round_circ_md_mr_mp_steane_Z_ML"

config = get_standard_config()
config["circuit"]["distances"] = distances
config["circuit"]["qec_rounds"] = qec_rounds
config["noise_model"]["noise_rates"] = noise_rates
config["decoder"]["type"] = "ml"
config["sampling"]["num_shots"] = num_shots_ML
config_ML = config

# write config to file
write_config(config_ML,folder_name_ML,backup=True)
```

```python
# Generate ML Data 
if True:
    config_path = os.path.join(PATH_TO_CONFIG_FOLDER, folder_name_ML + ".yaml")
    generate_new_data_from_config_file(config_path, folder_name_ML)
```

```python
# MWPM Config 
folder_name_MWPM = "multi_round_circ_md_mr_mp_steane_Z_MWPM"

config = get_standard_config()
config["circuit"]["distances"] = distances
config["circuit"]["qec_rounds"] = qec_rounds
config["noise_model"]["noise_rates"] = noise_rates
config["decoder"]["type"] = "mwpm"
config["sampling"]["num_shots"] = num_shots_MWPM
config_MWPM = config

# write config to file
write_config(config_MWPM,folder_name_MWPM,backup=True)
```

```python
# Generate MWPM Data 
if True:
    config_path = os.path.join(PATH_TO_CONFIG_FOLDER, folder_name_MWPM + ".yaml")
    generate_new_data_from_config_file(config_path, folder_name_MWPM)
```
